## Perth House Prices EDA and Visualization

### Importing modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Reading the dataset

```python
data = pd.read_csv('Perth House Prices.csv')
```

```python
data.head()
```

### Getting data info

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.columns
```

### Checking data for null values

```python
data.isnull().sum()
```

### Filling the missing values

```python
data['GARAGE'].replace('NaN',2.0, inplace=True)
```

```python
data.isnull().sum()
```

```python
data['BUILD_YEAR'].fillna(data['BUILD_YEAR'].mean(), inplace=True)
```

```python
data['BUILD_YEAR'] = data['BUILD_YEAR'].astype(int)
```

```python
data['BUILD_YEAR']
```

```python
data['DATE_SOLD'] = pd.Series(data['DATE_SOLD'], dtype="string")
```

```python
data['DATE_SOLD']
```

```python
sns.heatmap(data.isnull())
```

```python
data.columns
```

```python
data.drop('NEAREST_SCH_RANK', axis=1, inplace=True)
```

```python
data.head()
```

### Removing unwanted Columns

```python
data.drop(['POSTCODE', 'LATITUDE', 'LONGITUDE'],axis=1, inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

### Performing EDA on the data

### Top 10 Suburbs in Perth

```python
Suburbs = data['SUBURB'].value_counts().head(10)
```

```python
Suburbs
```

```python
Suburbs.values
```

```python
Suburbs.index
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=Suburbs.index, y=Suburbs.values)
plt.xlabel('Suburbs')
plt.title('Top 10 Suburbs in Perth')
plt.show()
```

### Top 10 Suburbs Price

```python
data.nlargest(10,['PRICE'])
```

### Price of house by bedrooms

```python
data['BEDROOMS'].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data['BEDROOMS'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=data['BEDROOMS'], y=data['PRICE'])
plt.xlabel('Bedrooms')
plt.ylabel('Price (in cr.)')
plt.title('Price of house by number of Bedrooms')
plt.show()
```

```python
sns.countplot(x=data['BATHROOMS'])
plt.show()
```
